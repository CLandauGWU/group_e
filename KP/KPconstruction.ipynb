```python
import geopandas as gpd
import os
ward = gpd.read_file('ward2012')
```

```python
ward.columns
```

```python
from shapely.geometry import Point
geometry = [Point(xy) for xy in zip(constr12_17.LONGITUDE.apply(float), constr12_17.LATITUDE.apply(float))]
crs = {'init': 'epsg:4326'}
points = gpd.GeoDataFrame(constr12_17, crs=crs, geometry=geometry)
fig, ax = plt.subplots()
ward.plot(ax=ax, color='red')
points.plot(ax=ax, color='black', marker='.', markersize=5)
ax.set_aspect('equal')
```

```python
geo_constr = gpd.sjoin(ward, points, how='left', op='intersects')
geo_constr.head()
```
